### Create a function named 'Graphs' which will take the following as arguments- 
#### data:  a dataframe of interest
#### variables: a list containing the names of the variables of your interest.
#### directory: a directory where you want the output to be stored. The default will be the working directory.

### The function, Graphs(data, variables, directory) will plot a histogram and a boxplot for all the numerical variables in the dataframe and a bar plot for all the categorical variables in the dataframe. The images of these plots will be exported as a png file in the given directory.

### Also, create a local package for the same which will help you to avoid this function in your notebook again and again.


```python
#import the requisite libraries
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np
import pandas as pd
#Change the directory to the default current working directory
os.chdir("C:\\Users\\92sup\\Downloads\\PYTHON")
# Read the dataset
df = pd.read_csv("cars.csv")
```

```python
#assigning the attributes dataset to variables
variables = df.columns
#df.info() 
#variables = ['MPG','Cylinders','Origin']
for var in variables:   #iterate through all the variables in the loop and to calculate the percentage of unique values.
    print(var,np.round(np.count_nonzero(df[var].unique())/len(df),2))
```

```python
df.info() # To know the detailed information about the data attributes
```

```python
#declare the function Graph that takes data,variables and the directory path
def graphs(file='cars.csv',variables=df.columns, directory ="C:\\Users\\92sup\\Downloads\\PYTHON" ):
    os.chdir(directory)  #change the directory 
    #df = pd.read_csv(file) 
    var_cont = []
    var_discrete = []
    for var in variables:
        if(df[var].dtype=='object'):   # 1st contition of seggregation of categorical and contineous variable
            var_discrete.append(var)   # append it into the list of discrete variables
        elif (np.count_nonzero(df[var].unique())/len(df)<0.05):  # 2nd condition for seggregation of categorical and contineous variable
            var_discrete.append(var)  # append it into the list of discrete variables
        else:
            var_cont.append(var)    #append it into the list of continuous variables
    print("Continuous", var_cont)
    print("Discrete", var_discrete)
    for var in var_cont:      # Loop for plotting graphs for continuous variables
        directory='C:\\Users\\92sup\\Downloads\\PYTHON\\GRAPH IMAGES\\Continuous'   #set directory to save the plots as png file to the desired folder
        os.chdir(directory)   # change directory
        fig, ax = plt.subplots(1,2, figsize=(16,8))
        sns.histplot(df[var],bins = 50, ax = ax[0])  #plot histogram by seaborn
        sns.boxplot(df[var], ax = ax[1])   # plot box plot
        plt.savefig(var)     # save the plot as image
        plt.show()
        
    for var in var_discrete:   # Loop for plotting graphs for discrete variables
        directory='C:\\Users\\92sup\\Downloads\\PYTHON\\GRAPH IMAGES\\Discrete'  # Set the ditrectory to save the plots at desired folder
        os.chdir(directory)    #change the directory
        view = df[var].value_counts().reset_index().rename(columns={'index':var,var:'count'})
        sns.barplot(x=var, y='count', data =view)  #plot the bar plots 
        plt.savefig(var)  # save the images
        plt.show()
        
```

```python
graphs()    #Call the function
```
